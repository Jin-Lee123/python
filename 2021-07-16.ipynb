```python
import turtle
import random

##함수 선언부분 ##
def screenLeftClick(x,y):
    global r, g, b
    r = random.random()
    g = random.random()
    b = random.random()
    turtle.pencolor((r,g,b))
    turtle.pendown()
    turtle.goto(x,y)
    
def screenRightClick(x,y):
    turtle.penup()
    turtle.goto(x,y)
    
def screenMidClick(x,y):
    global r, g, b
    tSize = random.randrange(1,10)
    turtle.shapesize(tSize)
    
## 변수 선언 부분 ##
pSize = 10
r,g,b = 0.0,0.0,0.0

## 메인 코드 부분 ##
turtle.title('거북이로 그림 그리기')
turtle.shape('turtle')
turtle.pensize(pSize)


turtle.onscreenclick(screenLeftClick,1)
turtle.onscreenclick(screenMidClick,2)
turtle.onscreenclick(screenRightClick,3)

turtle.done()

```

```python
import turtle, random

##전역 변수 선언 부분##
swidth,sheight,pSize, exitCount= 300,300,3,0
r,g,b,angle,dist,curX,curY = [0]*7
# 7 개의 변수 전부 초기화

##메인 코드 부분##
turtle.title('거북이가 맘대로 다니기')
turtle.shape('turtle')
turtle.pensize(pSize)
turtle.setup(width = swidth+30, height = sheight+30)
turtle.screensize(swidth,sheight)

while True :
    r = random.random()
    g = random.random()
    b = random.random()
    turtle.pencolor((r,g,b))
    # 색 랜덤

    angle = random.randint(0,360)
    dist  = random.randint(0,100)
    turtle.forward(dist)
    turtle.left(angle)
    curX = turtle.xcor()
    curY = turtle.ycor()
    
        
    if(-swidth/2 <= curX and curX <= swidth/2)and(-sheight/2 <= curY and curY <= sheight/2):
        pass
    else :
        turtle.penup()
        turtle.goto(0,0)
        turtle.pendown()
        
        exitCount+=1
        if exitCount >= 5 :
            break
        
turtle.done()
```

```python
import turtle as t

## 전역 변수 선언 부분 ##
swidth, sheight = 500, 500

## 메인 코드 부분 ##
t.title('무지개색 원그리기')
t.shape('turtle')
t.setup(width = swidth + 50, height = sheight + 50)
t.screensize(swidth, sheight)

t.penup()
t.goto(0,-250)
t.speed(10)
t.pendown()

for radius in range(1,250):
    color = ["red","orange","yellow","green","blue","navy","purple"]
    index = (radius-1) % 7
    t.pencolor(color[index])
    t.circle(radius)
    
t.done()
```

```python
import turtle as t, random 
from tkinter.simpledialog import *

##전역 변수 선언 부분##
inStr = ''
swidth,sheight = 500,500
tX,tY, txtSize = [0]*3

##메인 코드 부분##
t.title('거북이 글자쓰기')
t.setup(width=swidth+50, height=sheight+50)
t.screensize(swidth, sheight)
t.penup()
    
inStr = askstring('문자열 입력', "거북이 쓸 문자열을 입력")    

for ch in range(0,len(inStr)):
    tX = random.randrange(-swidth/2,swidth/2)
    tY = random.randrange(-sheight/2,sheight/2)
    
    #랜덤 위치
    r = random.random()
    g = random.random()
    b = random.random()
    t.pencolor((r,g,b))
    
    #랜덤 컬러
    txtSize = random.randrange(10, 50)
    
    #랜덤 사이즈
    t.goto(tX,tY)
    Str = inStr[ch]
    t.write(Str,font=('맑은고딕',txtSize,'bold'))
    
t.done()
```

```python
import turtle as t, random 
from tkinter.simpledialog import *
import math

##전역 변수 선언 부분##
inStr = ''
swidth,sheight = 500,500
tX,tY, txtSize = [0]*3

##메인 코드 부분##
t.title('거북이 글자쓰기')
t.setup(width=swidth+50, height=sheight+50)
t.screensize(swidth, sheight)
t.penup()
    
inStr = askstring('문자열 입력', "거북이 쓸 문자열을 입력")    

dist = 100 #반지름
angle = 0
value = int(360/len(inStr))


for ch in (inStr):
    rad = math.pi * angle / 180
    tX = math.cos(rad) * dist
    tY = math.sin(rad) * dist
    angle+=value
    #지정 위치 (원)
    
    r = random.random()
    g = random.random()
    b = random.random()
    t.pencolor((r,g,b))
    #랜덤 컬러
    txtSize = random.randrange(10, 50)
    #랜덤 사이즈
    t.goto(tX,tY)

    t.write(ch,font=('맑은고딕',txtSize,'bold'))
t.done()
```
